This is a sample example on how to use the scikit learn DBSCAN moduule. Part of the sample code is taken from scikit documentation example  <a href="http://scikit-learn.org/stable/auto_examples/cluster/plot_dbscan.html">here</a>.

```python
#let's import the modules we will use
from sklearn.cluster import DBSCAN
from sklearn.datasets.samples_generator import make_blobs
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np
```

We will first generate random data to cluster using DBSCAN. We will use a function from scikitlearn.datasets <a href= "http://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_blobs.html">make_blobs</a>. This function will generate isotropic Gaussian blobs of data based on the parameters that we specify. The parameters we will use:

<b>n_samples</b> which is the total number of points that we will generate. These points will be divided equally among the blobs/clusters.

<b>centers</b> which are the centers of each blob we will generate. As the blobs will be generated using Gaussian distribution, that means the points in each cluster will be spread away from the center we specify. 

<b>cluster_std</b> which is the standard deviation of the clusters.

<b>random_state</b> which is the random seed the random number generator will use.

```python
centers = [[1, 1], [-1, -1], [1, -1]]
X, labels_true = make_blobs(n_samples=750, centers=centers, 
                            cluster_std=0.4, random_state=0)
```

make_blobs will return the data generated, which in this case 750 2-dimensional points and the label of each generated point (in which cluster does that point belong to).

We will plot the data to see what they look like. 

```python
x,y = X.T
plt.scatter(x,y)
```

Now we will use StandardScaler to standarize the data points. This will basically trasform the data such that its distribution will have a mean value 0 and standard deviation of 1. 
If we plot the data after standarization in this case, we will see a small difference on the distribution as the data we generated are already of the same standard deviation on both dimensions.

```python
X = StandardScaler().fit_transform(X)
x,y = X.T
plt.scatter(x,y)
```

Standarizing the data is very important especially for models based on distance measurement and comparison like DBSCAN. Let's look at the following example. The first dimension has a range between 0 and 1 and the second dimension has a range between 0 and 10. 

```python
data = np.array([[0.1, 7], [0.2, 3], [0.6, 3], [0.4, 6]])

from sklearn.metrics.pairwise import euclidean_distances
print(euclidean_distances(data,data))
```

The above table shows the euclidean distance from every point in the data array to the other. Notice the second and third point has the same value in the second dimension (3) but the value in the first dimension has a big difference relative to the range of the dimension (0.2 and 0.6). Which result in a very small distance (0.4) relative to the other distances were the second dimension is dominating the distance measurement.
Now we will standarize the data points and calculate the distances again. 

```python
data = StandardScaler().fit_transform(data)
print(euclidean_distances(data,data))
```

The difference is obvious and the distances now give better values to compare and make desicions based on the distance between 2 points.

Now let's go back to clustering the data we generated above X.
We will build a DBSCAN model with epsilon = 0.3 and min number of sample points in cluster = 10.

```python
db = DBSCAN(eps=0.3, min_samples=10).fit(X)
```

We can find the label of each point by calling db.labels_. However, this will return an array of the labels of the 750 points. Instead we will use the unique function from numpy to find all the unique labels and then print the length of that array. Which should give us the number of clusters.

```python
unique_labels = np.unique(db.labels_)

print("number of clusters: ", len(unique_labels))
```

We are expecting 3 clusters, but we got 4. DBSCAN will give the outlier points label of -1. So let's find the value of the labels and the count of points for each label.

```python
unique, counts = np.unique(db.labels_, return_counts=True)
print(np.dstack((unique,counts)))
```

The unique function returns two arrays because we specified in the second parameter that we want the count to be returned. The count of every label will be stored in the counts array in the same order of the labels in the unique array. 
To change the two 1-dimensional arrays into one 2-dimensional array we use the numpy dstack function. 
The output shows us that we have 3 clusters and 18 outlier points labeled with -1.

To output the number of clusters, we shouldn't subtract one always from the number of clusters we calculated above. Because sometimes you might not have any outlier points and the label -1 might not exist. But we can check if -1 label exists and then subtract one.

```python
n_clusters_ = len(unique_labels) - (1 if -1 in unique_labels else 0)
print("number of clusters: ", n_clusters_)
```

```python
#the following plots the data with different colors for each cluster
#and black color for the outliers/noisy points
label_color_map = { -1 : 'k', 0 : 'r', 1 : 'b' , 2 : 'g'}
label_color = [label_color_map[l] for l in db.labels_]
x,y = X.T

plt.title('Estimated number of clusters: %d' % n_clusters_)
plt.scatter(x,y, c=label_color)
```
