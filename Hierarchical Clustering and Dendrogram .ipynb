This is a sample example on how to use the scipy dendrogram and linkage to visualize the hierarchical clustering and how to use sklearn to perform the clustering with specific number of clusters.

We will create 3 blobs of data points using `make_blobs` from `sklearn.datasets`.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn import datasets

centers = [[1, 1], [-1, -1], [1, -1]]
X, labels_true = datasets.make_blobs(n_samples=150, centers=centers, 
                            cluster_std=0.4, random_state=0)

x,y = X.T

plt.scatter(x,y)
plt.show()
```

From scipy will us `linkage` which will create the linkage matrix of dataset based on a linkage method that we pass to the function. We can also pass the distance metric which is euclidean by default.

```python
Z_ward = linkage(X, 'ward')
Z_complete = linkage(X, 'complete')
Z_average = linkage(X, 'average')
```

We created a linkage matrix using ward, complete and average methods. and now we will use `dendrogram` from scipy to plot the dendrogram and visualize the distribution of the hierarchy of the clusters.

```python
plt.title('Ward Hierarchical Clustering Dendrogram')
plt.xlabel('sample index')
plt.ylabel('distance')
dendrogram(
    Z_ward,
    leaf_rotation=90.,  # rotates the x axis labels
    leaf_font_size=10.,  # font size for the x axis labels
)
plt.show()


plt.title('Complete Hierarchical Clustering Dendrogram')
plt.xlabel('sample index')
plt.ylabel('distance')
dendrogram(
    Z_complete,
    leaf_rotation=90.,  # rotates the x axis labels
    leaf_font_size=10.,  # font size for the x axis labels
)
plt.show()

plt.title('Average Hierarchical Clustering Dendrogram')
plt.xlabel('sample index')
plt.ylabel('distance')
dendrogram(
    Z_average,
    leaf_rotation=90.,  # rotates the x axis labels
    leaf_font_size=10.,  # font size for the x axis labels
)
plt.show()
```

As we can see plotting the full hierarchy is messy, especially if we have a bigger dataset. We can truncate the dendrogram easily. One mode to do it in scipy is to plot the last `p` merged clusters. Which we can do by passing `'lastp'` to `truncate_mode` and then specifying the value of `p`. In this case we will have p = 12.

```python
plt.title('Ward Hierarchical Clustering Dendrogram')
plt.xlabel('sample index or (cluster size)')
plt.ylabel('distance')
dendrogram(
    Z_ward,
    truncate_mode='lastp',  # show only the last p merged clusters
    p=12,  # show only the last p merged clusters
    show_contracted=True,  # to get a distribution impression in truncated branches
    leaf_rotation=90.,  # rotates the x axis labels
    leaf_font_size=10.,  # font size for the x axis labels
)
plt.show()

plt.title('Complete Hierarchical Clustering Dendrogram')
plt.xlabel('sample index or (cluster size)')
plt.ylabel('distance')
dendrogram(
    Z_complete,
    truncate_mode='lastp',  # show only the last p merged clusters
    p=12,  # show only the last p merged clusters
    show_contracted=True,  # to get a distribution impression in truncated branches
    leaf_rotation=90.,  # rotates the x axis labels
    leaf_font_size=10.,  # font size for the x axis labels
)
plt.show()

plt.title('Average Hierarchical Clustering Dendrogram')
plt.xlabel('sample index or (cluster size)')
plt.ylabel('distance')
dendrogram(
    Z_average,
    truncate_mode='lastp',  # show only the last p merged clusters
    p=12,  # show only the last p merged clusters
    show_contracted=True,  # to get a distribution impression in truncated branches
    leaf_rotation=90.,  # rotates the x axis labels
    leaf_font_size=10.,  # font size for the x axis labels
)
plt.show()
```

Now this is better to look at. We can also see the size of every cluster of the `p` clusters. The height of the cluter hierarchy represnts the distance. From the dendrogram we can tell that we have 3 clusters with similar distribution of points inside each cluster as the 3 clusters are formed almost at the same height.

We can also notice that 2 of the 3 clusters are closer to each other. Which if we look at the dataset we generated it makes sense as the cluster on the top right is further away from the left bottom cluster than the bottom right cluster. 

And we can also guess that the 2 clusters that are closer to each other and getting merged first are the 2 bottom clusters as there are some noisy points in between the 2 clusters. 

Now we will use `AgglomerativeClustering` from sklearn to fit the model. Agglomerative clustering is bottom up hierarchical clustering. The other type of hierarchical clustering would start the algorithm with one cluster that contains all the points in the dataset and then keep on splitting up clusters until it forms a cluster for each point seperately. Which can be seen as top down form.

We will pass to the model the number of clusters we wish to get which we were able to determine from the dendrogram plot. And we will also pass the linkage method.

```python
from sklearn.cluster import AgglomerativeClustering
n_clusters = 3
average_linkage = AgglomerativeClustering(n_clusters=n_clusters,
                                          linkage="average").fit(X)
ward_linkage = AgglomerativeClustering(n_clusters=n_clusters,
                                      linkage="ward").fit(X)
complete_linkage = AgglomerativeClustering(n_clusters=n_clusters,
                                          linkage="complete").fit(X)
```

```python
label_color_map = {0 : 'g', 1 : 'r' , 2: 'c'}
label_color_average = [label_color_map[l] for l in average_linkage.labels_]
label_color_ward = [label_color_map[l] for l in ward_linkage.labels_]
label_color_complete = [label_color_map[l] for l in complete_linkage.labels_]

x,y = X.T

plt.title('ward linkage')
plt.scatter(x,y, c=label_color_ward)
plt.show()
plt.title('complete linkage')
plt.scatter(x,y, c=label_color_complete)
plt.show()
plt.title('average linkage')
plt.scatter(x,y, c=label_color_average)
plt.show()
```

By plotting the clusters we are able to see that all 3 methods almost performed the same way. We can plot the real labels from the dataset that we generated to compare. 

```python
label_color_map = {0 : 'g', 1 : 'r' , 2: 'c'}
label_color = [label_color_map[l] for l in labels_true]

plt.title('original dataset')
plt.scatter(x,y, c=label_color)
plt.show()
```

Although the colors are different, but we do not care about the order of the labels. If we compare closely we can notice that the average linkage gave the best results and it is only off by one data point.

However, this is not the best way to find out and compare results, especially many times we will not be able to visualize the dataset easily or accurately. 

We can use a score metric like completness score. The completness score checks if all the data points belonging to the same class from the true labels are formed into the same cluster from the generated clusters. A score of 1 means the clustering was perfect. 

```python
from sklearn.metrics.cluster import completeness_score
print("ward linkage score: ", 
      completeness_score(labels_true=labels_true, 
                         labels_pred=ward_linkage.labels_))
print("complete linkage score: ", 
      completeness_score(labels_true=labels_true, 
                         labels_pred=complete_linkage.labels_))
print("average linkage score: ", 
      completeness_score(labels_true=labels_true, 
                         labels_pred=average_linkage.labels_))
```

As we expected before, the average linkage gave a bit better results than ward and complete linkage methods.
